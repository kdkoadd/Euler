## Largest prime factor
### Problem 3
The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?

```python

class PrimeFactors():
    def __init__(self,number=1,primeTest=False):
        self.prime_factors=[]
        self.product = 1
        self.number = number
        self.stack = 0
        self.complete = False
        self.primeTest = primeTest # if true, return
    def reset(self,number,primeTest=False):
        self.prime_factors=[]
        self.product = 1
        self.number = number
        self.stack = 0
        self.complete = False
        self.primeTest = primeTest # if true, return
    def factors(self,n=0):
        if self.number == self.product:
            return self.prime_factors, True
        elif n == 0: #evaluate first prime pair (2,3)
            while self.number/self.product % 2 == 0:
                self.prime_factors.append(2)
                self.product *= 2
            while self.number/self.product % 3 == 0:
                self.prime_factors.append(3)
                self.product *= 3
        else:      
            while self.number/self.product % (6*n-1) == 0:
                self.prime_factors.append(6*n-1)
                self.product *= (6*n-1)
            while self.number/self.product % (6*n+1) == 0:
                self.prime_factors.append(6*n+1)
                self.product *= (6*n+1)

        #if primeTest == True, then just want to know if number is prime; so
        #as soon as self.prime_factors contains two or more elements, we know
        #number is not prime and can return False
        if self.primeTest and len(self.prime_factors) > 1:
            return None, False
        #recursion will overflow our stack - below (along with above while loops)
        #avoids recursion and stack overflow issue
        if self.stack == 1:
            self.stack = 0
            return self.prime_factors, False
        while not self.complete:
            self.stack = 1
            n += 1
            result, self.complete = self.factors(n)
            if result == None:
                return None, False

        return self.prime_factors, True
    
    def largest(self):
        self.factors(0)
        return max(self.prime_factors)
        
N = 13195
x = PrimeFactors(N)
print x.factors()
print "Largest prime factor of %i is %i" % (N,x.largest())
N = 600851475143
x = PrimeFactors(N)
print x.factors()
print "Largest prime factor of %i is %i" % (N,x.largest())
```

## Largest palindrome product
### Problem 4
A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99.

Find the largest palindrome made from the product of two 3-digit numbers.




```python
import math
palindromes = []
for x in range (100,1000):
    for y in range(x,1000):
        n = x*y
        #maybe avoid the necessity for the string compare to string reversal
        lsd = n % 10
        if (x % 10) * (y % 10) % 10 != lsd:
            continue
        l = math.floor(math.log10(n))
        msd = math.floor(n/10**l)
        if msd != lsd:
            continue
            
        #string compare to string reversal required
        if str(n) == str(n)[::-1]:
            palindromes.append(n)
            
print max(palindromes)
```

## 10001st prime
### Problem 7
By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.  

What is the 10 001st prime number?

```python
def nthPrime(n):
    pf = PrimeFactors()
    if n != int(n) or n <= 0:
        print "Invalid input: %f" % n
        return None
    if n > 0 and n < 3:
        return n+1
    else:
        p_count = 2
        x = 1
        while True:
            x += 1
            p_test = 2*x+1
            pf.reset(p_test,True)
            if not pf.factors()[1]:
                continue
            else:
                p_count += 1
                if p_count == n:
                    return p_test

N = 12
print "The %ith prime is %i" % (N,nthPrime(N))
N = 10001
print "The %ith prime is %i" % (N,nthPrime(N))
```

## Special Pythagorean triplet
### Problem 9
A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,

a^2 + b^2 = c^2  
For example, 3^2 + 4^2 = 9 + 16 = 25 = 5^2.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.

```python
#Euclid's formula for Pythagorean triplets:
#a = k(m^2 - n^2); b = k2mn; c = k(m^2 +n ^2); n < m; n,m integral
#subing into a+b+c=1000 -> k(m^2-n^2 + 2mn + m^2+n^2) = 1000
#simplified -> km(m+n) = 500
done = False
for n in range(1,501):
    if done:
        break
    for m in range(n+1,501):
        d = m*(m+n)
        if d > 500:
            continue
        k = 500/d
        if 500 % k == 0:
            a = k*(m**2-n**2)
            b = k*2*m*n
            c = k*(m**2+n**2)
            print "k: %i a: %i b: %i c: %i a+b+c: %i" % (k,a,b,c,a + b + c)
            print "a*b*c: %i" %(a*b*c)
            done = True
            break
```

## Summation of primes
### Problem 10
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

```python
#use The sieve of Eratosthenes to get the primes < n
#efficiency could be improved - fine for limits of problem (i.e., 2,000,000)
class SieveOfEratosthenes():
    def __init__(self,n):
        self.n = n
        self.primes = []
        self.not_primes = [False for x in range(0,n)]
    def getPrimes(self):
        return self.primes
    def generatePrimes(self):
        p = 2 # start with smallest prime
        while True:
            self.primes.append(p)
            for x in range (2*p,self.n,p):
                self.not_primes[x] = True
            while True:
                p += 1
                if p >= self.n:
                    return self.primes
                if self.not_primes[p]:
                    continue
                break
N = 2000000                
s = SieveOfEratosthenes(N)
print "The sum of all primes less than %i is: %i" % (N,sum(s.generatePrimes()))
```

## Highly divisible triangular number
### Problem 12
The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
#The canonical factorization of n is:
#p1^e1 . p2^e2 . ... . pk^ek; where pi are the prime factors of n: (2^2) * (7^1) for 28
#So, the number of divisors is
#tau(n) = (e1+1)(e2+1)(e3+1) ... (ek+1): (2+1)*(1+1) = 6 for 28
def triangleNumDivisors(n):
    pf = PrimeFactors()
    t_idx = 7
    while True:
        t_num = t_idx*(t_idx+1)/2 #get t_idx th triangle number
        pf.reset(t_num)
        pfs, _ = pf.factors() #get prime factors of t_num
        pfs_set = set(pfs) #get unique prime factors
        divisors = 1
        for p in pfs_set:
            divisors *= (pfs.count(p)+1)
        if divisors > n:
            return divisors, t_num
        t_idx += 1

divisors, triag_num = triangleNumDivisors(500)
print "There are %i divisors for triangle number: %i" % (divisors,triag_num)
```

## Longest Collatz sequence
### Problem 14
The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.



```python
def collatz(n,length=0):
    if n == 1: #assume Collatz is correct
        return length+1
    length += 1
    if n % 2 == 0:
        return collatz(n/2,length)
    else:
        return collatz(3*n + 1,length)
    
N = 1000000
max_length = 0
start_num = 1
for i in range(1,N):
    length = collatz(i)
    if length > max_length:
        start_num = i
        max_length = length
print "The length of the longest chain is: %i starting with n = %i" % (max_length,start_num)
```

## Lattice paths
### Problem 15
Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.


How many such routes are there through a 20×20 grid?

```python
#there will be (grid_size+1)^2 nodes in the grid
#e.g., for 2x2 lattice there will be 9 nodes (0...8)

def latticePaths(grid_size):
    grid_size += 1
    pos = 0
    nodes = (grid_size)**2
    move = [0 for x in range (0,nodes-1)] #number of moves from node 
    #add known moves to dictionary
    while pos < nodes - 1:
        if (pos+1) % grid_size == 0: #if can't move East, then only one move: South
            move[pos] = 1
        if pos + grid_size >= nodes: #if can't move South, then only one move: East
            move[pos] = 1
        pos += 1
    #work backwards from known moves to build moves from each node until reach
    #start node (i.e., node 0)
    for i in range(len(move)-1,-1,-1):
        if move[i] == 0: #node count not yet assigned
            move[i] = move[i+1] + move[i + grid_size]
    return move[0]

N = 20     
print "Number of paths through a %i-dimensional grid is: %i" % (N,latticePaths(N))        
```

## Number letter counts
### Problem 17
If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

```python
ones = {1:3,2:3,3:5,4:4,5:4,6:3,7:5,8:5,9:4}
teens = {10:3,11:6,12:6,13:8,14:8,15:7,16:7,17:9,18:8,19:8}
tens = {20:6,30:6,40:5,50:5,60:5,70:7,80:6,90:6}
hundreds = {100:10,200:10,300:12,400:11,500:11,600:10,700:12,800:12,900:11} 
thousands = {1000: 11}

N = 1000
letter_count = 0
for i in range(0,N+1):
    x = str(i)
    if len(x) == 4 and int(x[0])*1000 in thousands:
        letter_count += thousands[int(x[0])*1000]
        x = str(int(x) - 1000)
    if len(x) == 3 and int(x[0])*100 in hundreds:
        letter_count += hundreds[int(x[0])*100]
        x = str(int(x) - int(x[0])*100)
        if int(x) > 0:
            letter_count += 3
    if len(x) == 2 and i > 19 and int(x[0])*10 in tens:
        letter_count += tens[int(x[0])*10]
        x = str(int(x) - int(x[0])*10)
    if len(x) == 2 and int(x) in teens:
        letter_count += teens[int(x)]
        continue
    if len(x) == 1 and x[0] != '0':
        letter_count += ones[int(x[0])]
        
print "Letter count for numbers from 1 to %i (inclusive) is: %i" % (N,letter_count)           
```

Find the maximum total from top to bottom of the triangle below:

                  75

              20 04 82 47 65
             19 01 23 75 03 34
            88 02 77 73 07 63 67
           99 65 04 28 06 16 70 92

```python
nodes = """
75
95 64
17 47 82
18 35 87 10
20 04 82 47 65
19 01 23 75 03 34
88 02 77 73 07 63 67
99 65 04 28 06 16 70 92
41 41 26 56 83 40 80 70 33
41 48 72 33 47 32 37 16 94 29
53 71 44 65 25 43 91 52 97 51 14
70 11 33 28 77 73 17 78 39 68 17 57
91 71 52 38 17 14 91 43 58 50 27 29 48
63 66 04 68 89 53 67 30 73 16 69 87 40 31
04 62 98 27 23 09 70 98 73 93 38 53 60 04 23
"""
def largestPath(nodes):
    #compute number of nodes 
    levels = [x.split() for x in nodes.split('\n') if x != '']
    node_count = len(levels)*(len(levels)+1)/2 #assumes triangle arrangement of input
    print "There are %i levels/rows with a total of %i nodes in the input triangle." % (len(levels),node_count)
    #initialize cost list (i.e., to node cost)
    partial_obj_val = [[0 for y in range(0,len(x.split()))] for x in nodes.split('\n') if x != '']
    #build triangle levels, converting string to int
    arc_cost = [[int(i) for i in x.split()] for x in nodes.split('\n') if x != '']
    partial_obj_val[0][0] = arc_cost[0][0]

    #longest path problem - graph is acyclic and directed
    current_level = 0
    next_level = 1
    while current_level <= len(levels) - 2: #start at top level and work down
        for i in range (0,len(arc_cost[current_level])): #nodes at the current level 
            next_level = current_level + 1 #level below current level, i
            from_cost = partial_obj_val[current_level][i] #cumulative cost to reach current_level node i
            for j in range(i,i+2): #loop through nodes at level next
                partial_obj_val[next_level][j] = max(arc_cost[next_level][j]+from_cost,partial_obj_val[next_level][j])
        current_level += 1
    return max([max(y) for y in partial_obj_val])

max_path_val = largestPath(nodes)
print "The value of the maximum path through the triangle is:",max_path_val
```

## Counting Sundays
### Problem 19
You are given the following information, but you may prefer to do some research for yourself.

1 Jan 1900 was a Monday.  
Thirty days has September,
April, June and November.
All the rest have thirty-one,
Saving February alone,
Which has twenty-eight, rain or shine.
And on leap years, twenty-nine.  
A leap year occurs on any year evenly divisible by 4, but not on a century unless it is divisible by 400.  

How many Sundays fell on the first of the month during the twentieth century (1 Jan 1901 to 31 Dec 2000)?

```python
#easiest to use calendar API
import calendar

sunday_count = 0
for year in range(1901,2001):
    for month in range (1,13):
        cal = calendar.monthcalendar(year, month)
        week_one = cal[0]
        sunday = week_one[calendar.SUNDAY]
        if sunday == 1:
            sunday_count += 1

print "%i Sundays fell on the first of the month during the 20th century." % sunday_count
```

## Factorial digit sum
### Problem 20
n! means n × (n − 1) × ... × 3 × 2 × 1

For example, 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800,
and the sum of the digits in the number 10! is 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27.

Find the sum of the digits in the number 100!




```python
'''
class Factorial returns N! as a list of digits by performing digit-wise
multiplication to avoid overflow issues - mimicks hand multiplication
'''
class Factorial():
    def __init__(self):
        self.result = [1] #list of digits in n!
        self.n = 0
    def factorial(self,n):
        self.result = [1] #list of digits in n!
        if n in [0,1]:
            return [1]
        self.n = n
        for x in range (2,n+1):
            self.__multiply(x) #2 x 3 x 4 x ... x n
        return [self.result[i] for i in range(len(self.result)-1,-1,-1)] #reorder
    def asInt(self,n):
        l = self.factorial(n)
        return int(''.join(str(x) for x in l))
    def __multiply(self,x):
        carry = 0
        #multiply digits (stored as LSD to MSD) of result with x
        for i in range (0,len(self.result)):
            product = self.result[i] * x + carry
            self.result[i] = product % 10 #get LSD of product and store
            carry  = product / 10
        #Put carry in result least significant digits first 
        while carry:
            self.result.append(carry % 10)
            carry = carry / 10
N = 100
f = Factorial()
print "The sum of the digits in %i! is %i" % (N,sum(f.factorial(N)))
```

## Amicable numbers
### Problem 21
Let d(n) be defined as the sum of proper divisors of n (numbers less than n which divide evenly into n).
If d(a) = b and d(b) = a, where a ≠ b, then a and b are an amicable pair and each of a and b are called amicable numbers.

For example, the proper divisors of 220 are 1, 2, 4, 5, 10, 11, 20, 22, 44, 55 and 110; therefore d(220) = 284. The proper divisors of 284 are 1, 2, 4, 71 and 142; so d(284) = 220.

Evaluate the sum of all the amicable numbers under 10000.




```python
import math
def properDivisors(n):
    divs = [1]
    upper = int(math.sqrt(n))
    if n > 1 and n == upper*upper:
        divs.append(upper)
    else:
        upper += 1
    for div in range(2,upper):
        if n % div == 0:
            divs.append(div)
            divs.append(n/div)
    return divs

def amicable(n):
    divs_of_n = properDivisors(n)
    d_of_divs_of_n = sum(divs_of_n)
    d_of_d_of_divs_of_n = sum(properDivisors(d_of_divs_of_n))
    if d_of_divs_of_n != n and d_of_d_of_divs_of_n == n:
        return d_of_d_of_divs_of_n + d_of_divs_of_n
    else:
        return 0

N = 10000
total = 0
for i in range (1, N):
    total += amicable(i)
print "The sum of all the amicable numbers under %i is %i" % (N,total/2)
```

## Non-abundant sums
### Problem 23
A perfect number is a number for which the sum of its proper divisors is exactly equal to the number. For example, the sum of the proper divisors of 28 would be 1 + 2 + 4 + 7 + 14 = 28, which means that 28 is a perfect number.

A number n is called deficient if the sum of its proper divisors is less than n and it is called abundant if this sum exceeds n.

As 12 is the smallest abundant number, 1 + 2 + 3 + 4 + 6 = 16, the smallest number that can be written as the sum of two abundant numbers is 24. By mathematical analysis, it can be shown that all integers greater than 28123 can be written as the sum of two abundant numbers. However, this upper limit cannot be reduced any further by analysis even though it is known that the greatest number that cannot be expressed as the sum of two abundant numbers is less than this limit.

Find the sum of all the positive integers which cannot be written as the sum of two abundant numbers.




```python
import math
def properDivisors(n):
    divs = [1]
    upper = int(math.sqrt(n))
    if n > 1 and n == upper*upper:
        divs.append(upper)
    else:
        upper += 1
    for div in range(2,upper):
        if n % div == 0:
            divs.append(div)
            divs.append(n/div)
    return divs

class AbundantNumbers():
    def __init__(self,max_num):
        self.max_num = max_num
        #create list of abundant numbers below max_num
        self.abundant_numbers = set([x for x in range(12,28124) if sum(properDivisors(x)) > x])
        self.not_abundant_sum = []
    def getNotAbundantSum(self):
        return self.not_abundant_sum
    def getAbundantNumbers(self):
        return self.abundant_numbers
    '''
    To determine if a number n is the sum of two abundant numbers, search for
    some i,j in abundant_numbers where i < n and j = n-i
    '''
    def isAbundantSum(self,n):
        for i in self.abundant_numbers:
            if i >= n: #n can't be sum of two pos nums, one of which is equal to or larger than n
                return False
            j = n - i
            if j in self.abundant_numbers: #clearly, i + n - i = n and i and j are both abundant nums
                return True
        return False
    def notAbundantSum(self):
        self.not_abundant_sum = [n for n in range(1,self.max_num) if not self.isAbundantSum(n)]
        return self.not_abundant_sum

N = 28124
abundant = AbundantNumbers(N)
print "The sum of pos ints which cannot be written as sum of two abundant nums:",sum(abundant.notAbundantSum())
```

F1 = 1
